### Aggregating tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values

```python
# Get the absolute values
D = tf.constant([-7, -10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Let's go through the following forms aggregation:
* Get the minimum
* Get the maximun
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Creating a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
np.min(E)
```

```python
# find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the variance of our tensor
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# Find the standard deviation
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

```python
#Find the variance other methode
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

***Find the positional maximum and minimum***




```python
# Create a new tensor for finding positional max and min
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
np.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check of equalities
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional min
tf.argmin(F)
```

```python
# Find the min using the positional mini index
F[tf.argmin(F)]
```
