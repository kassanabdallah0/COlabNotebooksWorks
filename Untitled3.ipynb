```python
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from math import sqrt 
from skimage.color import rgb2gray
import glob
from skimage.io import imread
import numpy as np
import cv2
from skimage import color
import time
```

```python
t0 = time.time()
file_name = "/content/172.jpg"
im= imread(file_name, as_gray= True)
blobs_log = blob_log(im, max_sigma=30, num_sigma=10, threshold= 0.1)
blobs_log[: 2] = blobs_log[: 2] * sqrt(2)
fig, ax = plt.subplots(1,1)
plt.imshow(im, cmap= plt.cm.gray)

for blob in blobs_log:
  y, x, r = blob
  c = plt.Circle((x,y), r + 3, color = 'lime', linewidth = 1, fill = False)
  ax.add_patch(c)
plt.show()
t1 = time.time()
total = t1 - t0
num_rows = len(blobs_log)
print("nombre des impact", num_rows)
print(total)
```

```python
fig, ax = plt.subplots(1,1)
plt.imshow(im, cmap= plt.cm.gray)
t00 = time.time()
img = cv2.imread("/content/172.jpg", 0)
blur = cv2.cv2.GaussianBlur(img, (3, 3), 0)
ret, thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
contours1, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img, contours1, 0, (0, 255, 0), 1)
t11 = time.time()
total1 = t11 - t00
ax.imshow(img)
plt.show()
print(len(contours1))
print(total1)
```
