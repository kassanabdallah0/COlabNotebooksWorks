```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
  print(0)
i = 0
while True:
  ret, frame = cap.read()
  frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
  frame = cv2.resize(frame, (150, 150))
  if cv2.waitKey(1):
    if ord('s'):
      cv2.imwrite('/content/plaque/'f'{i}.jpg')
      i += 1
    if ord('q'):
      break
cap.release()
cv2.destroyAllWindows()
```
