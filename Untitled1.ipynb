```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
blank = np.zeros((1920, 1080, 3), dtype=np.uint8)
blank = blank + 255
```

```python
#cv2_imshow(blank)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1
color = (255, 0, 0)
thickness = 2
org = (50, 50)
blank = cv2.putText(blank, 'OpenCV', org, font, 
                   fontScale, color, thickness, cv2.LDR_SIZE)
```

```python
cv2_imshow(blank)
```
