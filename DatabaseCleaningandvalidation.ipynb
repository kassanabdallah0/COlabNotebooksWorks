**Static Dropdown Lists:**

These can be directly identified from the Options_Map JSON file. Any key with a non-empty list as its value represents a static dropdown list.

```python
# Static Dropdown Lists Mapping
static_dropdowns = {key: value for key, value in options_map.items() if value}
```

**Dynamic Dropdown Lists:**

Any key in the Options_Map JSON file with an empty list as its value represents a dynamic dropdown list.
The actual values for these dynamic dropdown lists can be found in the other JSON files, where the keys in those files correspond to some values in the Options_Map.

```python
# Resetting the dynamic_dropdowns dictionary to ensure it's empty
dynamic_dropdowns = {key: {} for key in options_map if not options_map[key]}
# Populating the dynamic_dropdowns dictionary
for key in dynamic_dropdowns.keys():
    for filename, content in dynamic_dropdown_data.items():
        if key in content:
            dynamic_dropdowns[key].update(content[key])
```

* Static_dropdowns: Contains all the static dropdown lists from the Options_Map JSON file.
* Dynamic_dropdowns: Contains all the dynamic dropdown lists. The keys are the headers from the Excel sheet (which are also keys in Options_Map), and the values are dictionaries that map values from Options_Map to their corresponding dynamic dropdown lists.

# Data transformation

1. Reading through the Excel data column by column.
2. For each column, check if the header exists in the options_map dictionary.
3. If it does, it means this column has dropdown list values.
4. For each value in the column:
  * If the header has a static dropdown list (in static_dropdowns), check if the value matches any in the list.
  * If the header has a dynamic dropdown list (in dynamic_dropdowns), find the appropriate dynamic list based on the associated static dropdown value and check if the Excel value matches any in the dynamic list.
5. Replace or transform the Excel values based on the dropdown lists.

```python
# Iterate through each column in the dataframe
for column in df.columns:
    # Check if the column is in the static dropdowns
    if column in static_dropdowns:
        # Replace values in the column based on the dropdown list
        df[column] = df[column].apply(lambda x: x if x in static_dropdowns[column] else 'NaN')

    # Check if the column is in the dynamic dropdowns
    elif column in dynamic_dropdowns:
        # Here, you'll need logic to determine which dynamic dropdown list to use
        # This depends on the associated static dropdown value
        for key, sub_dict in dynamic_dropdowns[column].items():
            df[column] = df[column].apply(lambda x: x if x in sub_dict else 'NaN')
```

**1. Fuzzy Matching:**

Fuzzy matching can be useful when there are slight discrepancies between the values in your Excel sheet and the dropdown lists, such as typos, variations in spelling, or different wordings.



```python
!pip install fuzzywuzzy
from fuzzywuzzy import process
```

```python
# Fuzzy Matching Function
def get_best_match(value, choices, threshold=80):
    best_match, score = process.extractOne(value, choices)
    if score >= threshold:
        return best_match
    return None  # Return None if no match above the threshold
```

**2. Semantic Similarity:**

Semantic similarity can be used to identify words or phrases that have similar meanings, even if they aren't lexically similar.

```python
!pip install spacy
!python -m spacy download en_core_web_md
import spacy
```

```python
!python -m spacy download fr_core_news_md
```

```python
nlp = spacy.load('fr_core_news_md')
```

```python
# Semantic Similarity Function
def get_semantic_match(value, choices, threshold=0.8):
    value_vec = nlp(value)
    best_match = max(choices, key=lambda choice: value_vec.similarity(nlp(choice)))
    similarity = value_vec.similarity(nlp(best_match))
    if similarity >= threshold:
        return best_match
    return None  # Return None if no match above the threshold
```

```python
# Combined Matching Function
def get_combined_match(value, choices):
    fuzzy_match = get_best_match(value, choices)
    if fuzzy_match:
        return fuzzy_match
    return get_semantic_match(value, choices) or value  # Fall back to the original value if no match found
```

```python
def load_dynamic_dropdown_data(uploaded_dynamic_dropdowns):
    dynamic_dropdown_data = {}
    for filename in uploaded_dynamic_dropdowns.keys():
        with open(filename, 'r') as file:
            content = json.load(file)
            # Extracting the nested dictionary
            dynamic_dropdowns = list(content.values())[0]
            # Merging with the main dynamic_dropdown_data
            dynamic_dropdown_data.update(dynamic_dropdowns)
    return dynamic_dropdown_data
```

```python
def find_related_dropdowns(options_map, dynamic_dropdown_data):
    related_dropdowns = {}
    static_keys = {val: key for key, values in options_map.items() for val in values if values}

    for dynamic_key, dynamic_values in dynamic_dropdown_data.items():
        static_key = static_keys.get(dynamic_key)
        if static_key:
            dynamic_key_in_options_map = next((key for key, values in options_map.items() if not values), None)
            if dynamic_key_in_options_map:
                related_dropdowns[dynamic_key_in_options_map] = static_key
                print(f"Found related static and dynamic dropdowns: {dynamic_key_in_options_map} (dynamic) -> {static_key} (static)")
    return related_dropdowns
```

```python
# Load the dynamic dropdown data
dynamic_dropdown_data = load_dynamic_dropdown_data(uploaded_dynamic_dropdowns)

# Find related static and dynamic dropdowns
related_dropdowns = find_related_dropdowns(options_map, dynamic_dropdown_data)

# Print the related dropdowns
print("Related Dropdowns:", related_dropdowns)
```

```python
def transform_data(df, options_map, dynamic_dropdowns):
    # Identify related static and dynamic dropdowns
    related_dropdowns = find_related_dropdowns(options_map, dynamic_dropdowns)
    print(f"Related Dropdowns: {related_dropdowns}")

    # Transform static dropdown lists
    for column in df.columns:
        if column in options_map and options_map[column]:
            print(f"Transforming static dropdown list: {column}")
            # Apply matching for static dropdowns (you can replace with your matching logic)
            df[column] = df[column].apply(lambda x: get_combined_match(x, options_map[column]))

    # Transform dynamic dropdown lists
    for column, related_static in related_dropdowns.items():
        print(f"Transforming dynamic dropdown list: {column} (related to {related_static})")
        for idx, row in df.iterrows():
            # Find the corresponding static value
            associated_static_value = row[related_static]

            # Find the corresponding dynamic list
            dynamic_list = dynamic_dropdowns.get(associated_static_value, {})
            if dynamic_list:
                # Apply matching for dynamic dropdowns (you can replace with your matching logic)
                matched_value = get_combined_match(row[column], dynamic_list)
                df.at[idx, column] = matched_value
                print(f"Matched value: {row[column]} -> {matched_value}")
```
